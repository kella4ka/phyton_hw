```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Данные о машинах
auto = pd.read_excel('C:\\Users\\KulkovaAV\\Desktop\\Phyton\\Auto.xlsx')
tt = auto.dropna()
tt.head()
```

```python
#По данному графику можно определить выбросы. Также, по графику видно минимальное и максимальные значения расходов, видно, 
#что 25% выбирают машину с расходом выше 13,5 л. 
tt['Расход'].plot(kind='box')
```

```python
#Чтобы посмотреть попарную взаимосвязь между всеми переменными в dataframe, используем функцию pairplot 
sns.pairplot(tt[['Расход', 'Стоимость']])
```

```python
#График гистограммы позволяет сделать первичную оценку выборки на принадлежность к какому-либо распределению. 
#Например, расход очень напоминает нормальное распределение
tt.hist(figsize=(10,10))
```

```python
#Посмотреть на гистограмму и плотность распределения можно с помощью данной функции
sns.distplot(tt['Расход'].dropna())
```

```python
#Чтобы понять, связаны ли линейно переменные x и y можно построить scatter совместо с линией тренда
fig, ax = plt.subplots(figsize=(5,5))
sns.regplot(x="Расход", y="Стоимость", data=tt)
```
