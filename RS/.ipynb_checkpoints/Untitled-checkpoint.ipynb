```python
import pandas as pd
import numpy as np
from datetime import datetime

from tqdm import tqdm_notebook

from sklearn.feature_extraction.text import TfidfTransformer, CountVectorizer
from sklearn.neighbors import NearestNeighbors

import matplotlib.pyplot as plt
%matplotlib inline
```

1. Использовать dataset MovieLens
2. Построить рекомендации (регрессия, предсказываем оценку) на фичах:
3. TF-IDF на тегах и жанрах
4. Средние оценки (+ median, variance, etc.) пользователя и фильма
5. Оценить RMSE на тестовой выборке

```python
links = pd.read_csv('links.csv')
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
tags = pd.read_csv('tags.csv')
```

```python
movies_with_tags = pd.merge(movies, tags, how = 'left', on = 'movieId')
```

```python
movies_with_tags_with_ratings = pd.merge(movies_with_tags, ratings, how = 'inner', left_on = ['userId', 'movieId'], right_on =  ['userId', 'movieId'])
```

```python
movies_with_tags_with_ratings
```

```python
year_tag = []
year_rating = [] 

for t1, t2 in zip(movies_with_tags_with_ratings.timestamp_x.values,movies_with_tags_with_ratings.timestamp_y.values):
    d1 = datetime.fromtimestamp(t1)
    d2 = datetime.fromtimestamp(t2)
    year_tag.append(str(d1.year))
    year_rating.append(str(d2.year))
```

```python
movies_with_tags_with_ratings['year_tag'] = np.array(year_tag)
movies_with_tags_with_ratings['year_rating'] = np.array(year_rating)
```

```python
movies_with_tags_with_ratings
```

```python
movies_with_tags_with_ratings.groupby(by = ['userId']).movieId.count().plot(kind="bar") 
```

```python
#Средняя оценка по тегу
tag_groups = movies_with_tags_with_ratings.groupby(['title', 'tag']).agg({'rating':['count', 'mean']}).reset_index()
tag_groups.columns = ['title', 'tag', 'r_count', 'r_mean']
tag_groups.sort_values(by=['r_count'])
```

```python
#Средняя оценка по фильму
muvies_groups = movies_with_tags_with_ratings.groupby(['movieId']).agg({'rating':['min','max', 'count', 'mean']}).reset_index()
muvies_groups.columns = ['movieId', 'rm_min', 'rm_max', 'rm_count', 'rm_mean']
#muvies_groups.sort_values(by=['movieId', 'rm_count'])
```

```python
mtr_ = pd.merge(movies_with_tags_with_ratings, muvies_groups, how = 'left', on = 'movieId')
```

```python
#Средний рейтинг, который ставит каждый пользователь
ratings_groups = movies_with_tags_with_ratings.groupby(['userId']).agg({'rating':['min','max', 'count', 'mean']}).reset_index()
ratings_groups.columns = ['userId', 'ru_min', 'ru_max', 'ru_count', 'ru_mean']
#ratings_groups.sort_values(by=['userId', 'ru_count'])
```

```python
mtr_with_mean = pd.merge(mtr_, ratings_groups, how = 'left', on = 'userId')
```

```python
def change_string(s):
    return ' '.join(s.replace(' ', '').replace('-', '').split('|'))
```

```python
mtr_with_mean['genres'] = mtr_with_mean['genres'].apply(change_string)
```

```python
mtr_with_mean['feature'] = mtr_with_mean['genres'] + ' ' + mtr_with_mean['tag']
```

```python
mtr_with_mean
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import NearestNeighbors

from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
# Регрессия TF-IDF на тегах и жанрах
X = mtr_with_mean['feature']
y = mtr_with_mean['rating']
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.2)
```

```python
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(X_train)
```

```python
tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)
```

```python
model = RandomForestRegressor()
model.fit(X_train_tfidf, y_train)
```

```python
predict = count_vect.transform(X_test)
X_tfidf2 = tfidf_transformer.transform(predict)

y_pred = model.predict(X_tfidf2)
```

```python
df = pd.DataFrame({'Film': mtr_with_mean[mtr_with_mean.index.isin(y_test.index)]['title'].values, 'DESCR': mtr_with_mean[mtr_with_mean.index.isin(y_test.index)]['feature'].values, 'Actual': y_test, 'Predicted': y_pred})
df
```

```python
print('MAE', mean_absolute_error(y_test, y_pred))
print('MSE', mean_squared_error(y_test, y_pred))
```

```python
df.groupby(['Actual', 'Film', 'DESCR']).agg({'Predicted': 'mean' }).reset_index()
```

```python
# Регрессия на средних оценках пользователя и фильма
X_train, X_test, y_train, y_test = train_test_split(mtr_with_mean[['rm_mean', 'ru_mean']], mtr_with_mean['rating'], test_size=0.2)
rf = RandomForestRegressor()
rf.fit(X_train, y_train)
y_rf = rf.predict(X_test)
```

```python
print('R2', r2_score(y_test, y_rf))
print('MAE', mean_absolute_error(y_test, y_rf))
print('MSE', mean_squared_error(y_test, y_rf))
```
