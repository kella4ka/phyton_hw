```python
def salary(a):
    if a == 'low':
        return 0
    if a == 'medium':
        return 1
    if a == 'high':
        return 2
def sales(a):
    if a == 'sales':
        return 0
    if a == 'accounting':
        return 1
    if a == 'hr':
        return 2
    if a == 'technical':
        return 3
    if a == 'support':
        return 4
    if a == 'management':
        return 5
    if a == 'IT':
        return 6
    if a == 'product_mng':
        return 7
    if a == 'marketing':
        return 8
    if a == 'RandD':
        return 9
    
    

df['salary_num'] = df['salary'].apply(salary)
df['sales_num'] = df['sales'].apply(sales)
df.head()
```

           проверить взаимосвязи между несколькими переменными c использованием t-test статистики

```python
#pvalue < 0.05 => выборки имеют разные средние (т.е. группы статистически отличаются)
out_ = df2[df2['left']==0]['satisfaction_level']
in_ = df2[df2['left']==1]['satisfaction_level']

stats.ttest_ind(out_[:len(in_)], in_)
```

```python
#pvalue > 0.05 => не имеется статистическое различие между уровнем удовлетворенности  IT и product_mng 
IT = df2[df2['sales_num']==6]['satisfaction_level']
product_mng = df2[df2['sales_num']==7]['satisfaction_level']
stats.ttest_ind(IT[:len(product_mng)], product_mng)
```

```python
sns.distplot(IT)
sns.distplot(product_mng)
plt.title('Плотность вероятности распределения satisfaction_level')
plt.legend(['IT', 'product_mng'])
plt.show()
```

```python
#pvalue > 0.05 => не имеется статистическое различие между уровнем удовлетворенности marketing и генеральной совокупности в целом. 
female_bps = df2[df2['sales_num'] == 8]['satisfaction_level']
stats.ttest_1samp(female_bps, df['satisfaction_level'].mean())
```

                 Проверить несколько гипотез для данных, чтобы понять почему сотрудники покидают компанию

```python
headers = list(df2.columns.values)
```

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(-1,1))

np_scaled = min_max_scaler.fit_transform(df2)
df_norm = pd.DataFrame(np_scaled)
df_norm.columns = headers
df_norm.head()
```

```python
print('Количество оставшихся сотрудников: ', len(df2[df2['left']==0]))
print('Количество уволившихся сотрудников: ', len(df2[df2['left']==1]))
```

```python
count_classes = pd.value_counts(df2['left'], sort = True).sort_index()
count_classes.plot(kind = 'bar')
```

```python
#Выбираю из общей таблицы dataframe, соответствующий только сотрудникам, которые покинули компанию
df3 = df_norm[df_norm['left']==1]
df3.head()
```

```python
#Хочу посмотреть на уровень удовлетворенности сотрудника, относительно следующих факторов input_x.
output_y = df3['satisfaction_level']
input_x = df3[['salary_num','time_spend_company','average_montly_hours','promotion_last_5years','Work_accident', 'number_project']]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

y = output_y
salary = input_x['salary_num']
time_spend_comp = input_x['time_spend_company']
avg_montly_hours = input_x['average_montly_hours']
prom = input_x['promotion_last_5years']
accident = input_x['Work_accident']
number_project = input_x['number_project']

x = np.vstack([np.array(salary), np.array(time_spend_comp), np.array(avg_montly_hours), 
               np.array(prom), np.array(accident), np.array(number_project)]).T

X_train, X_test, y_train, y_test = train_test_split(x,
                                                 y,
                                                 test_size = 0.8,
                                                 random_state = 3)

model = LinearRegression().fit(X_train, y_train)
model.score(X_test, y_test)
```

Получается, что уровень удовлетворенности на 58% состоит из перечисленных факторов. Каждый из этих факторов вносит свой вклад на уровень удовлетвоеренности сотрудника. Т.о. чем меньше удовлетворен сотрудник, тем больше вероятность, что он уволится.
